### List Comprehensions


```python
squares = []

for i in range(1,10001): #itera evalua 1 por 1
    squares.append(i**2)
    
print(squares)
```

```python
# hay proceso paralelizado: lo hace al mismo tiempo todo, no 1 por 1. Hacer eso ahorra tiempo y capacidad computacional
my_list = [x for x in range (1,11)]
print(my_list)
```

```python
my_list = [(x**2 + 15) for x in range (1,11)]
print(my_list)
```

```python
my_list = [(x**2 + 15) for x in range (1,11) if x%5==0 ]
print(my_list)
```

```python
my_list = [(x**2 + 15) for x in range (1,11) if x%5==0 and x*10>50] # esta sentencia te puede servir para muchas cosas, es más practico, es list conprension
print(my_list)
```

```python
squares = [value**2 for value in range(1, 11)]
print(squares)
```

### Métodos en listas

Cómo modificar elementos de una lista

```python
a[2] = 1
print(a)
```

<b>append</b>

Añade un elemento al final de la lista.

```python
a = [1, 2]
a.append(3)
print(a)
```

<b>extend</b>

Permite añadir una lista a la lista inicial.

```python
a = [1, 2]
a.extend([3, 4])
print(a)
```

<b>insert</b>

Añade un elemento en una posición o índice determinado.

```python
a = [1, 3]
a.insert(1, 10)
print(a) 
```

<b>remove</b>

Elimina elementos de una lista

```python
a = [1, 2, 3]
a.remove(2)
print(a) 
```

<b>pop</b>

Elimina por defecto el último elemento de la lista, pero si se pasa como parámetro un índice permite borrar elementos diferentes al último

```python
a = [1, 2, 3]
a.pop()
print(a)
```

<b>reverse</b>
Inverte el orden de la lista

```python
a = [1, 2, 3]
a.reverse()
print(a)
```

<b>sort</b>

Ordena los elementos de menor a mayor.


```python
a = [3, 1, 2]
a.sort()
print(a) 
```

```python
a = [3, 1, 2]
a.sort(reverse = True)
print(a) 
```

### Agregando nuevos pares Key-Value

```python
alien_0 = {'color': 'green', 'points': 5}
print(alien_0)

alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)
```

### Comenzando con un diccionario vacio

```python
alien_0 = {}

alien_0['color'] = 'green'
alien_0['points'] = 5

print(alien_0)
```

### Modificando valores de un diccionario

```python
alien_0 = {'color': 'verde'}
print(f"El alien es {alien_0['color']}.")

alien_0['color'] = 'amarillo'
print(f"El alien ahora es {alien_0['color']}.")
```

```python
alien_0 = {'x_position': 0, 'y_position': 25, 'speed': 'medium'}
print(f"Original position: {alien_0['x_position']}")

if alien_0['speed'] == 'slow':
  x_increment = 1
elif alien_0['speed'] == 'medium':
  x_increment = 2
else:
  x_increment = 3

alien_0['x_position'] = alien_0['x_position'] + x_increment
print(f"New position: {alien_0['x_position']}")
```

### Removiendo pares Key-Value

```python
alien_0 = {'color': 'green', 'points': 5}
print(alien_0)

del alien_0['points']
print(alien_0)
```

### Lista de diccionarios

```python
alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
  print(alien)
```

```python
aliens = []

# Make 30 green aliens.
for alien_number in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)

# Show the first 5 aliens.
for alien in aliens[:5]:
    print(alien)
print("...")

# Show how many aliens have been created.
print(f"Total number of aliens: {len(aliens)}")
```

```python
aliens = []

# Make 30 green aliens.
for alien_number in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)

for alien in aliens[:3]:
    if alien['color'] == 'green':
        alien['color'] = 'yellow'
        alien['speed'] = 'medium'
        alien['points'] = 10

# Show the first 5 aliens.
for alien in aliens[:5]:
    print(alien)
print("...")

# Show how many aliens have been created.
print(f"Total number of aliens: {len(aliens)}")
```

```python
aliens = []

# Make 30 green aliens.
for alien_number in range(30):
  new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
  aliens.append(new_alien)

for alien in aliens[:3]:
  if alien['color'] == 'green':
    alien['color'] = 'yellow'
    alien['speed'] = 'medium'
    alien['points'] = 10
        
  elif alien['color'] == 'yellow':
    alien['color'] = 'red'
    alien['speed'] = 'fast'
    alien['points'] = 15

# Show the first 5 aliens.
for alien in aliens[:5]:
    print(alien)
print("...")

# Show how many aliens have been created.
print(f"Total number of aliens: {len(aliens)}")
```

### Una lista en un diccionario

```python
# Store information about a pizza being ordered.
pizza = {
         'crust': 'thick',
         'toppings': ['mushrooms', 'extra cheese'],
        }

# Summarize the order.
print(f"You ordered a {pizza['crust']}-crust pizza "
  "with the following toppings:")

for topping in pizza['toppings']:
  print(f"\t{topping}")
```

```python
favorite_languages = {
       'jen': ['python', 'ruby'],
       'sarah': ['c'],
       'edward': ['ruby', 'go'],
       'phil': ['python', 'haskell'],
       }

for name, languages in favorite_languages.items():
  print(f"\n{name}'s favorite languages are:")

  for language in languages:
    print(f"\t{language}")
```
